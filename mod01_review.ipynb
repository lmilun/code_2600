### Find the mean, median, and mode of the feature "Age."

```python
# mean
sum = 0
for a in Credit.Age:
    sum += a

print(sum / len(Credit.Age))
```

```python
# median
Ages = Credit.Age
Ages = Ages.sort_values()
if len(Ages) % 2 == 0:
    print((Ages[len(Ages) / 2] + Ages[len(Ages) / 2 - 1]) / 2)
else:
    print(Ages[len(Ages) / 2 - 0.5])
```

```python
# mode
mode = [[-1000],-1000]
cur = [-1000,0]

for a in Ages:
    if a != cur[0]:
        if cur[1] > mode[1]:
            mode[0] = [cur[0]]
            mode[1] = cur[1]
        elif cur[1] == mode[1]:
            mode[0].append(cur[0])
        cur[0] = a
        cur[1] = 1
    else:
        cur[1] += 1

if cur[1] > mode[1]:
    mode[0] = cur[0]
    mode[1] = cur[1]
elif cur[1] == mode :
    mode[0].append(cur[0])

print(mode[0])
```

### Create a subset of Credit that only contains the columns Age, Region, and Income. Call it "Subset."

```python
Subset = Credit[['Age', 'Region', 'Income']]
Subset
```

### From this subset, create two partitions: 
- One that contains records for Ages below 50, and one that contains Ages 50+
- Call them "Younger" and "Older"

```python
Younger = Subset[Subset['Age'] < 50]
Younger
```

```python
Older = Subset[Subset['Age'] >= 50]
Older
```

### Use these partitions to compare the mean and median Income for each age group (less than 50 vs. 50+)

```python
# mean of younger
sum = 0
for a in Younger.Age:
    sum += a

print(sum / len(Younger.Age))
```

```python
# mean of older
sum = 0
for a in Older.Age:
    sum += a

print(sum / len(Older.Age))
```

```python
# median of younger
yAges = Younger.Age
yAges = yAges.sort_values()
if len(yAges) % 2 == 0:
    print((yAges[len(yAges) / 2] + yAges[len(yAges) / 2 - 1]) / 2)
else:
    print(yAges[len(yAges) / 2 - 0.5])
```

```python
# median of older
oAges = Older.Age
oAges = oAges.sort_values()
if len(oAges) % 2 == 0:
    print((oAges[len(oAges) / 2] + oAges[len(oAges) / 2 - 1]) / 2)
else:
    print(oAges[len(oAges) / 2 - 0.5])
```

### What does the difference between mean and median suggest about the distribution of income?

The difference between the mean and median in the younger ages is small. This suggests that the data is not skewed, and centered around 39.

The difference between the mean and median in the older ages is large, with the mean being smaller than the median. This suggests that the distribution of income is skewed to the right.

Often, banks will use a variable called _utilization_ to assess credit risk.
Utilization is the ratio of balance to limit (i.e., what % of the credit limit is being used?)
### Add a feature to the Credit dataset to measure utilization

```python
Credit['Utilization'] = Credit['Balance'] / Credit['Limit']
Credit['Utilization']
```

### What is the maximum utilization value in this set? 

```python
max = -10
for a in Credit.Utilization:
    if a > max:
        max = a

print(max)
```
